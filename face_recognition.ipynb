```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('/home/ishaan/Downloads/face.jpg')
print type(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
face_cascade=cv2.CascadeClassifier('/home/ishaan/Codes/Datasets/face_cascades/haarcascade_frontalface_default.xml')
```

```python
face=face_cascade.detectMultiScale(img)
print face
```

```python
x=face[0][0]
y=face[0][1]
w=face[0][2]
z=face[0][3]

found_face=cv2.rectangle(img,(x,y),(x+w,y+z),(255,0,0),2)
plt.imshow(found_face)
plt.show()
```

```python
eyes_cascade=cv2.CascadeClassifier('/home/ishaan/Codes/Datasets/face_cascades/haarcascade_eye.xml')
```

```python
eyes=eyes_cascade.detectMultiScale(found_face)
```

```python
print type(eyes)
```

```python
print eyes
```

```python
eye_1_x=eyes[0][0]
eye_1_y=eyes[0][1]
eye_1_w=eyes[0][2]
eye_1_z=eyes[0][3]
eye_2_x=eyes[1][0]
eye_2_y=eyes[1][1]
eye_2_w=eyes[1][2]
eye_2_z=eyes[1][3]


eye_rectangle=cv2.rectangle(found_face,(eye_1_x,eye_1_y),(eye_1_x+eye_1_w,eye_1_y+eye_1_z),(0,255,0),2)
eye2_rectangle=cv2.rectangle(eye_rectangle,(eye_2_x,eye_2_y),(eye_2_x+eye_2_w,eye_2_y+eye_2_z),(0,255,0),2)

plt.imshow(eye2_rectangle)
plt.show()
```
