```python
strings=""
dir_path="/home/ishaan/Codes/Datasets/ego_hands/_LABELLED_SAMPLES/"

sort_dir_path=sorted(os.listdir(dir_path))
for dir_s in sort_dir_path:
    image_dir_path=dir_path+str(dir_s)
    try:
        cur_pol_mat=sc.loadmat(image_dir_path+"/"+"polygons.mat")
        #extract info of every frame from polygon.mat 
        image_no=0
        sort_image_dir_path=sorted(os.listdir(image_dir_path))
        for images in sort_image_dir_path:
                image=cv2.imread(image_dir_path+"/"+str(images))
                list_cord_2d=_ext_x_y(cur_pol_mat,image_no)
                hds_cord_2d=np.array(list_cord_2d,dtype=float)
                image_no=image_no+1
     
                if(str(images)=="polygons.mat"):
                    continue
                else:
                    image_shape=image.shape
                    image_shrt_path="_LABELLED_SAMPLES/"+str(dir_s)+"/"+str(images)
                    ego_hands.write(image_shrt_path+str(image_shape)+"\n")
                    
                    rows=hds_cord_2d.shape[0]
                    cols=hds_cord_2d.shape[1]
                    
                    for cur_hd in range(rows):
                        for cur_cord in range (cols):
                            string=str(hds_cord_2d[cur_hd][cur_cord])+","
                            strings=strings+string
                            
                        strings=strings[:-1]
                        ego_hands.write(strings+";")
                        strings=""
                    ego_hands.write("\n")
                    #ego_hands.write(image_shrt_path+" "+str(image_shape)+"\n"+str(x)+","+str(y)+","+str(w)+","+str(h)+";";"+"\n")
                
    except Exception as e:
        print e
                

ego_hands.close()
                
```
